```python
# Tranformando minha coluna em datas 
df['PrazoAtendimento'] = pd.to_datetime(df['PrazoAtendimento'], format='%d/%m/%Y', errors='coerce')
```

```python
# Tranformando minha coluna em datas 
df['DataResposta'] = pd.to_datetime(df['DataResposta'], format='%d/%m/%Y', errors='coerce')
```

```python
# Tranformando minha coluna em datas 
df['DataRegistro'] = pd.to_datetime(df['DataRegistro'], format='%d/%m/%Y', errors='coerce')
```

```python
# Vendo como está
df.head()
```

```python
# Mostrando quantidade de nulos em cada coluna
df.isna().sum()
```

```python
# Apagando dados nulos apartir da coluna "DataResposta"
df = df.dropna(axis=0, subset=['DataResposta'])
```

```python
# Subtraindo os dias
df['Dias_demora-parcial'] = df['DataResposta'] - df['PrazoAtendimento']
```

```python
# Vendo como está
df.head()
```

```python
# Transformando a coluna em int
df['Dias_demora'] = df['Dias_demora-parcial'].dt.days
```

```python
# Deixando apenas os pedidos que demoraram mais de 30 dias para ser respondidos
df_descumpridas = df[df.Dias_demora >0]
```

```python
# Ordenando do mais demorado para o menos demorado
df_ordenado = df_descumpridas.sort_values(by=['Dias_demora'], ascending=False)
```

```python
# Vendo como está
df_ordenado.head(10)
```

```python
#ordenando pelos anos com mais registros de pedidos atrasados
df_ordenado.groupby([df_ordenado['DataRegistro'].dt.year]).size()
```
